```python
covid_dict=dict(covid19.getAll())
```

```python
covid_dict['locations'][1]['id']
```

```python
dict={'Id':[],'Country':[],'Country code':[],'Country population':[],'Province':[],
      'Last updated':[],'Latitude':[],'Longitude':[],'Confirmed':[],'Deaths':[],'Recovered':[]}
for i in range(263):
    dict['Id'].append(covid_dict['locations'][i]['id'])
    dict['Country'].append(covid_dict['locations'][i]['country'])
    dict['Country code'].append(covid_dict['locations'][i]['country_code'])
    dict['Country population'].append(covid_dict['locations'][i]['country_population'])
    dict['Province'].append(covid_dict['locations'][i]['province'])
    dict['Last updated'].append(covid_dict['locations'][i]['last_updated'])
    dict['Latitude'].append(covid_dict['locations'][i]['coordinates']['latitude'])
    dict['Longitude'].append(covid_dict['locations'][i]['coordinates']['longitude'])
    dict['Confirmed'].append(covid_dict['locations'][i]['latest']['confirmed'])
    dict['Deaths'].append(covid_dict['locations'][i]['latest']['deaths'])
    dict['Recovered'].append(covid_dict['locations'][i]['latest']['recovered'])
```

```python
covid_df=pd.DataFrame(dict)
```

```python
covid_df.set_index('Id', inplace=True)
```
